```python

def detect_outliers_iqr(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

# Get numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for column_name in numerical_cols:
    outliers = detect_outliers_iqr(df[column_name])
    print(f"Outliers in {column_name}: {outliers}")

    # Visualizing outliers (optional)
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column_name])
    plt.title(f'Boxplot of {column_name}')
    plt.show()
    outlier_mask = (df[column_name] < detect_outliers_iqr(df[column_name]).min()) | (df[column_name] > detect_outliers_iqr(df[column_name]).max())
    df = df[~outlier_mask]
```
